```python
import matplotlib
from matplotlib import pyplot as plt
```

```python
x = [9,12,15]
y = [10,1,6]

plt.plot(x,y)

plt.title('Demo Plotting')
plt.ylabel('Y axis')
plt.xlabel('X axis')

#Showing what we plotted
plt.show()
```

```python
from matplotlib import style
```

```python
#For line and width changes
style.use('ggplot')
```

```python
x = [9,12,15]
y = [10,1,6]

x2 = [16,4,1]
y2 = [6,5,2]

# can plot specifically, after just showing the defaults:
plt.plot(x,y,linewidth=5)
plt.plot(x2,y2,linewidth=5)

plt.title('Multiple lines demo USING ggplot style')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
plt.plot(x,y,'g',label='line one', linewidth=5)
plt.plot(x2,y2,'c',label='line two',linewidth=5)

plt.title('Demo with legend')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.legend()

plt.grid(True,color='k')

plt.show()
```

```python
#Plot Bar charts
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

x = [15,18,19]
y = [1,6,16]

x2 = [2,8,1]
y2 = [4,15,8]


plt.bar(x, y, align='center')

plt.bar(x2, y2, color='g', align='center')


plt.title('Demo Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
#Scatter plots
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

x = [15,18,19]
y = [1,6,16]

x2 = [2,8,1]
y2 = [4,15,8]

plt.scatter(x, y)#, align='center')

plt.scatter(x2, y2, color='g')#, align='center')


plt.title('Demo Scatter plot')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```
